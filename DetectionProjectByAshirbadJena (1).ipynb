```python
from ultralytics import YOLO
from PIL import Image
import cv2

model = YOLO("yolov8n.pt")
im1 = Image.open("/content/maxresdefault.jpeg")
results = model.predict(source=im1, save=True)
im2 = cv2.imread("/content/maxresdefault.jpeg")
results = model.predict(source=im2, save=True, save_txt=True)

results = model.predict(source=[im1, im2])
```

```python
def invoice_provided(results):
  for result in results[0].names.values():
    if "invoice" in result:
      return True
  return False

if 'person' in results[0].names.values():
  if invoice_provided(results):
    print("No alert needed. Invoice provided.")
  else:
    print("Alert generated. Invoice not provided.")
else:
  print("No person detected with the invoice")
```
